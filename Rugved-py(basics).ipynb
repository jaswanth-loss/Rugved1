```python
 #Define a function named “triple_and” that takes three parameters and returns
#True only if they are all True and False otherwise


def triple_and(a,b,c) :
    return a and b and c
def main() :
    a = input("Enter value of a (true/false): ").lower() == "true"
    b = input("Enter value of b (true/false): ").lower() == "true"
    c = input("Enter value of c (true/false): ").lower() == "true"
    print(triple_and(a,b,c))
main()
```

```python
""" Write a python program to sort a string alphabetically and print the count of each
character."""
s = input("Enter string: ")

sorted_s = ''.join(sorted(s))
print("Sorted string:", sorted_s)

print("Character count:")
for ch in sorted(set(s)):
    print(ch, ":", s.count(ch))
```

```python
def is_hill_number(n):
    s = str(n)
    i = 1

    while i < len(s) and s[i] > s[i-1]:
        i += 1

    if i == 1 or i == len(s):
        return False

    while i < len(s) and s[i] < s[i-1]:
        i += 1

    return i == len(s)

num = int(input("Enter number: "))
print(is_hill_number(num))
```

```python
def selection_sort_string(s):
    arr = list(s)
    n = len(arr)

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return ''.join(arr)

print(selection_sort_string("python"))
```

```python
#. Find the factorial of a given number using recursion.
def factorial(n) :
    if n == 0 or n==1 :
        return 1
    else :
        return n * factorial(n-1)
def main() :
    n=int(input("Enter a number: "))
    print(factorial(n))
main()
```

```python
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)

print(is_anagram("listen", "silent"))  # True
```

```python
s = input("Enter string: ")
n = int(input("Enter n: "))

if len(s) % n != 0:
    print("Error: Division not possible")
else:
    parts = [s[i:i+n] for i in range(0, len(s), n)]
    if all(part == parts[0] for part in parts):
        for part in parts:
            print(part)
    else:
        print("Error: Sequences are not same")
```

```python
def caesar_cipher(text, shift):
    result = ""

    for ch in text:
        if ch.isalpha():
            base = ord('A') if ch.isupper() else ord('a')
            result += chr((ord(ch) - base + shift) % 26 + base)
        else:
            result += ch

    return result

print(caesar_cipher("Hello World", 3))  # Khoor Zruog
```
